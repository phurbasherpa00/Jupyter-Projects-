### Creating a Combo Chart to find out the Percent Attrition based on Education Levels.

```python
df.value_counts("Attrition_Flag") 
```

```python
df.value_counts("Education_Level")
```

### Step 1 : Creating smaller table to store information we need.

```python
df_education = df[["Attrition_Flag", "Education_Level"]].copy() 
```

```python
df_education.head()
```

### Step 2 : Assigning Binary number and converting Existing Customer : 0 and Attrited Customer : 1

```python
 df_education["Attrition_Flag_binary"] = df_education["Attrition_Flag"].map({"Existing Customer": 0, "Attrited Customer": 1})
```

```python
df_education.head()
```

### Step 3 : Using Group by to sum and count.
> * Group by + sum will not affect the number 0
> * Group by + Count will include 0 when being counted.


```python
df_attrition_number = df_education.groupby("Education_Level").sum()
```

```python
df_total_number = df_education.groupby("Education_Level").count()
```

```python
df_attrition_number
```

```python
df_total_number
```

```python
df_total_number.rename({"Attrition_count": "Attrition_total_count"}, axis = 1, inplace = True) 
```

```python
df_total_number
```

### Step 4 : Calculating the rate and the percentage

```python
df_attrition_rate = df_attrition_number["Attrition_count"] / df_total_number["Attrition_total_count"]
```

```python
df_attrition_rate
```

```python

df_education_final = pd.concat([df_total_number,df_attrition_rate], axis = 1)
```

```python
df_education_final
```

```python
df_education_final.columns
```

```python

df_education_final.rename({0: "Attrition_float"}, axis = 1,inplace = True)
```

```python
df_education_final["Attrition_pct"] = df_education_final["Attrition_float"].map("{:.2%}".format)
```

```python
df_education_final
```

### Step 5 : Plotting the Data Frame that was created.

```python
x = df_education_final.index
y1 = df_education_final["Attrition_total_count"]
y2 = df_education_final["Attrition_float"]
y3 = df_education_final["Attrition_pct"]
```

### This Combo chart of line and bar that shows Attrition Percentage for Different Education Levels
> * Line chart should show the Attrition Percentage
> * Bar chart should show the Total count.

```python
fig, ax = plt.subplots(figsize = (15,9))
# drawing the bar chart
ax.bar(x, y1,
color = "orange", edgecolor = "blue", label = "Total Count")
ax.set_xticks(x, x, fontsize = 20, rotation = 45) # create a twin x so we have the right y axis
ax1 = ax.twinx()
# useing y2 the float to plot
ax1.plot(x, y2,
linewidth = 3, linestyle = ":", color = "magenta", marker = 'o', markersize = 15, markerfacecolor = "cyan", label = "Percentage of attrition")
# using y2 as value, y3 as ticks
ax1.set_yticks(y2, y3, fontsize = 20)
# plot the legend
ax.legend(loc = "upper left", fontsize = 20) 
ax1.legend(loc = "upper right", fontsize = 20)

plt.show()
```

```python
df_ori = pd.read_csv("BankChurners.csv")
```

```python
df_ori["Attrition_Binary"] = df_ori["Attrition_Flag"].map({"Existing Customer": 0,
                                                          "Attrited Customer": 1})
```

```python
df_ori.head()
```

```python
df_ori.value_counts("Education_Level")
```

```python
df_high = df_ori[df_ori["Education_Level"] == "High School"]
```

```python
required_column = ['CLIENTNUM','Credit_Limit','Customer_Age','Total_Revolving_Bal',
                   'Total_Trans_Ct', "Attrition_Binary"]
```

```python
df_high_num = df_high[required_column]
```

```python
df_high_num.head()
```

```python
df_high_num.head(2)
```

```python
df_high_num["Age_std"] = df_high_num["Customer_Age"].rolling(window = 30).std()
df_high_num["Credit_Limit_std"] = df_high_num["Credit_Limit"].rolling(window = 30).std()
df_high_num["Total_Revolving_Bal_std"] = df_high_num["Total_Revolving_Bal"].rolling(window = 30).std()
df_high_num["Total_Trans_Ct_std"] = df_high_num["Total_Trans_Ct"].rolling(window = 30).std()
```

```python
df_high_std = df_high_num.dropna()
```

```python
df_high_std.head()
```

```python
df_high_std.reset_index(inplace = True)
```

```python
df_high_std.drop(columns = "index", inplace = True)
```

```python
df_high_std_1 = df_high_std[df_high_std["Attrition_Binary"] == 1]
```

```python
df_high_std_0 = df_high_std[df_high_std["Attrition_Binary"] == 0]
```

```python
df_high_std_0.value_counts("Attrition_Binary")
```

```python
df_high_std_clean = pd.concat([df_high_std_0.iloc[:1400], df_high_std_1, df_high_std_0.iloc[1401:]])
```

```python
#reset the index because the concat one's index are mixed
df_high_std_clean.reset_index(inplace = True)
```

```python
df_high_std_clean["Attrition_Binary"].plot()
```

```python
df_high_std_clean.head()
```

```python
dependent = df_high_std_clean["Attrition_Binary"]
```

```python
independent = df_high_std_clean[['Age_std','Credit_Limit_std','Total_Revolving_Bal_std', 'Total_Trans_Ct_std']]
```

```python
independent_new = sm.add_constant(independent)
```

```python
my_model = sm.OLS(dependent, independent_new).fit()
```

```python
my_model.summary()
```

```python
prediction = my_model.predict(independent_new)
```

```python
plt.plot(df_high_std_clean.index, prediction, color = 'red')
```

```python
fig, ax = plt.subplots()
ax.plot(df_high_std_clean.index, df_high_std_clean.Age_std, label = "age std")
ax.plot(df_high_std_clean.index, df_high_std_clean.Credit_Limit_std / 1000, label = "credit_limit_std(in 1000 dollars)")
ax.plot(df_high_std_clean.index, df_high_std_clean.Total_Revolving_Bal_std/100, label = "Total_Revolving_Bal_std(in 100 dollars)")
ax.plot(df_high_std_clean.index, df_high_std_clean.Total_Trans_Ct_std, label = "Total_trans_Ct std")
ax.set_ylim(0, 50)
ax.legend(loc = "upper left")

ax1 = ax.twinx()
ax1.plot(df_high_std_clean.index, df_high_std_clean.Attrition_Binary, color = "black", label = "Binary Attrition")
ax1.plot(df_high_std_clean.index, prediction, color = 'cyan', label = "prediction")
ax1.legend()
```
